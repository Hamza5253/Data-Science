```python
#importing useful libraries
import os
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import sklearn.preprocessing as skp 
import sklearn.model_selection as skm 

#import classification modules 

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import (accuracy_score, confusion_matrix, precision_score, recall_score,
roc_auc_score, roc_curve, auc, f1_score)
```

```python
data = pd.read_csv("creditriskmodeling.csv")
print("Shape: ", data.shape,"\n")
print("Data Types: \n", data.dtypes)
```

```python
data.isnull().sum()
```

```python
data.dropna(axis="rows", inplace=True)
```

```python
data.isnull().sum()
```

```python
y = data['Default'].copy()
```

```python
x = data.drop('Default', axis="columns")
```

```python
#standardize all numerical columns to the same range
x = skp.StandardScaler().fit(x).transform(x)
x
```

```python
y = np.where(y=='Yes',1,0)
```

```python
y
```

```python
#Create Train-test dataset split from manual split 

trainX, testX, trainY, testY = skm.train_test_split(x, y, test_size=0.25, random_state=53)
print("shape of train split: ",trainX.shape,trainY.shape)
print("shape of test split: ",testX.shape,testY.shape)
```

```python
# NB on Manual Split
clf = GaussianNB()
clf.fit(trainX,trainY)
prediction = clf.predict(testX)
print("Accruracy of NB(%): \n",accuracy_score(testY, prediction)*100)
fpr, tpr, _ =roc_curve(testY,prediction)
print("AUC of NB (%): \n", auc(fpr,tpr)*100) 
```

```python
#create arrays for storing 
accarray = []
aucarray = []

#Define the split into 4 folds 

kf = skm.StratifiedKFold(n_splits=5, random_state=99, shuffle =False)
print(list(kf.split(x,y)))
```

<b> 
In above result there are 4 tuples and in every tuple there are to arrays which represents the train test splits.
kf.split() function shows give the row indexes of splitted data.</b>

```python
for train_index, test_index in kf.split(x,y):
    trainX,testX = x[train_index], x[test_index]
    trainY, testY = y[train_index],y[test_index]
    
    clf = KNeighborsClassifier(n_neighbors = 5)
    clf.fit(trainX,trainY)
    prediction = clf.predict(testX)
    accarray= np.append(accarray,accuracy_score(testY,prediction)*100)
print("Average accuracy of K-Nearest Neighbor(%):", np.mean(accarray))
```
